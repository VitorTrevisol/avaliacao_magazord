```python
import os
from pymongo import MongoClient
from sqlalchemy import create_engine
import pandas as pd
```

```python
mongo_uri = os.getenv('MONGO_URI')
mongo_db = os.getenv('MONGO_DB')
pg_uri = os.getenv('POSTGRES_URI')
```

```python
db_mongo = MongoClient(mongo_uri)
engine_pg = create_engine(pg_uri)
print("✅ Conexões configuradas via variáveis de ambiente!")
```

```python
# 3. Selecionar o banco de dados definido no .env
db = db_mongo[mongo_db]

# 4. Listar e imprimir as coleções (tabelas)
colecoes = db.list_collection_names()

if not colecoes:
    print(f"O banco '{mongo_db}' está vazio ou ainda não tem coleções criadas.")
else:
    print(f"Coleções encontradas no banco '{mongo_db}':")
    for col in colecoes:
        print(f" - {col}")
```

```python
# 2. Lista das coleções que você quer visualizar
colecoes = ['users', 'carts', 'products']

for nome_col in colecoes:
    print(f"\n--- Conteúdo da Coleção: {nome_col} ---")
    
    # Busca todos os documentos da coleção
    documentos = list(db[nome_col].find())
    
    if len(documentos) == 0:
        print(f"A coleção '{nome_col}' está vazia.")
    else:
        # Converte para DataFrame para uma visualização bonita no Jupyter
        df = pd.DataFrame(documentos)
        
        # Opcional: remover a coluna '_id' do MongoDB para facilitar a leitura
        if '_id' in df.columns:
            df = df.drop(columns=['_id'])
            
        display(df.head()) # Mostra as primeiras 5 linhas de cada coleção
```

```python
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv

# 1. Carregar credenciais
load_dotenv()
postgres_uri = os.getenv('POSTGRES_URI')
engine = create_engine(postgres_uri)

# 2. Lista de tabelas que criamos no pipeline
tabelas = [
    'dim_users', 'dim_locations', 'dim_company', 'dim_finance',
    'dim_products', 'dim_brands', 'fact_sales', 'fact_sales_items'
]

# 3. Dicionário para armazenar os DataFrames
dw_data = {}

print("📥 Carregando tabelas do Data Warehouse...")

for tabela in tabelas:
    try:
        # Lemos a tabela inteira para um DataFrame
        dw_data[tabela] = pd.read_sql(f'SELECT * FROM "{tabela}"', engine)
        print(f"✅ {tabela}: {len(dw_data[tabela])} registros carregados.")
    except Exception as e:
        print(f"❌ Erro ao carregar {tabela}: {e}")

# --- EXEMPLOS DE USO ---

# Para ver os usuários com seus endereços (JOIN em memória com Pandas)
df_completo = dw_data['dim_users'].merge(dw_data['dim_locations'], on='location_id')
display(df_completo.head())

# Para ver o faturamento total por marca
df_vendas = dw_data['fact_sales_items'].merge(dw_data['dim_products'], on='product_id')
df_vendas = df_vendas.merge(dw_data['dim_brands'], on='brand_id')
faturamento_marca = df_vendas.groupby('brand_name')['total_price'].sum().sort_values(ascending=False)

print("\n💰 Faturamento por Marca:")
print(faturamento_marca)
```
